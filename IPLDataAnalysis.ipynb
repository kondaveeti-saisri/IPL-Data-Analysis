```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv(r"V:\sai\dsprojects\ipl\matches.csv")
```

```python
# Display the first few rows
print(df.head())
```

```python
# Display dataset info
print(df.info())
```

```python
# Check for missing values
print("Missing Values:\n", df.isnull().sum())

# Standardize column names
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Fill or drop missing values (temporary approach)
df.fillna({'city': 'Unknown'}, inplace=True)
  # Replace missing cities with "Unknown"
# Drop 'method' column only if it exists
if 'method' in df.columns:
    df.drop(columns=['method'], inplace=True)
   # Too many missing values, dropping this column

# Re-check dataset info
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count wins for each team
team_wins = df['winner'].value_counts()

# Plot a bar chart
plt.figure(figsize=(10,5))
sns.barplot(x=team_wins.index, y=team_wins.values, hue=team_wins.index, palette="viridis", legend=False)
plt.xticks(rotation=90)
plt.title("Number of Matches Won by Each Team")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a new column to check if toss winner is also match winner
df['toss_win_match_win'] = df['toss_winner'] == df['winner']

# Count matches where toss winner also won the match
toss_effect = df['toss_win_match_win'].value_counts()

# Plot bar chart
plt.figure(figsize=(6,4))
sns.barplot(x=toss_effect.index, y=toss_effect.values, hue=toss_effect.index, palette="coolwarm", legend=False)
plt.xticks(ticks=[0,1], labels=["Lost After Toss Win", "Won After Toss Win"])
plt.title("Does Winning the Toss Help in Winning the Match?")
plt.ylabel("Number of Matches")
plt.show()

```

```python
# Count Player of the Match awards
top_players = df['player_of_match'].value_counts().head(10)

# Plot a bar chart
plt.figure(figsize=(10,5))
sns.barplot(x=top_players.index, y=top_players.values, hue=top_players.index, palette="magma", legend=False)
plt.xticks(rotation=90)
plt.title("Top 10 Players with Most 'Player of the Match' Awards")
plt.xlabel("Players")
plt.ylabel("Number of Awards")
plt.show()
```

```python
# Select numerical columns for correlation analysis
numeric_df = df[['result_margin', 'target_runs', 'target_overs']]

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Match Statistics")
plt.show()
```
